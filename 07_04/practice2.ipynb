```python
class Person:
    def __init__(self,name):
        #建立一個attribute
        self.name = name
    def introduce(self):
        print(f"哈囉，我的名字是{self.name}")
```

```python
p2 = Person("robert")
p2.introduce()
```

```python
p3 = Person("jenny")
p3.introduce()
```
