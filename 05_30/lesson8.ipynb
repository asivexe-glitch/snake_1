```python
file = open("student.txt", 'r', encoding='UTF-8')
print(type(file))
content = file.read()
print(content)
file.close()
file.closed
```

```python
with open("student.txt", 'r', encoding='UTF-8') as file:
    content = file.read()

print(file.closed)  # True，因為 with 區塊結束後會自動關閉檔案
```
