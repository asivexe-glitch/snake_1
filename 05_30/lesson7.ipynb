```python
# open(file, mode='r', buffering=-1, encoding=None, errors=None, newline=None, closefd=True, opener=None)
```

```python
try:
	with open("student.txt", "r", encoding="utf-8") as file:
			data = file.read()
			print(data)
except Exception as e:
	print("發生錯誤:", e)
```

```python
# open("student.txt") 引數值的呼叫

try:
	with open("student.txt") as file:
			data = file.read()
			print(data)
except Exception as e:
	print("發生錯誤:", e)
```

```python
# open(file="student.txt") 引數名稱的呼叫，可以不依順序，只要依照名稱

try:
	with open(file="student.txt", encoding='utf-8') as file:
			data = file.read()
			print(data)
except Exception as e:
	print("發生錯誤:", e)
```

```python
# open("student.txt") 引數值的呼叫，必須依順序

try:
	with open("student.txt", 'r', -1, None, 'UTF-8') as file:
			data = file.read()
			print(data)
except Exception as e:
	print("發生錯誤:", e)
```

```python
# 混合名稱的呼叫：前面用引數值(要依順序)，後面用引數名稱(免按順序)

try:
	with open("student.txt", encoding='utf-8', mode='r') as file:
			data = file.read()
			print(data)
except Exception as e:
	print("發生錯誤:", e)
```
