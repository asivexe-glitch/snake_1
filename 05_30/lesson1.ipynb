```python
# 資料結構
# 程式結構

# list
# 可以修改
scores = [89,92,76,91,77]
print(scores)
```

```python
# tuple
# tuple建立後不可修改
# 暫時儲存用
scores1 = [85,92,87,73,59]
print(scores1)
print(scores1[0])
scores1[0] = 100
```

```python
# dictionary
students = {'chinese':89, 'english':94, 'history':92, 'discover':95}
print(students['english'])
```

```python
for s in scores1:
    print(s)
```

```python
# dictionary
students = {'chinese':89, 'english':94, 'history':92, 'discover':95}

for key in students:
    print(students[key])
```
